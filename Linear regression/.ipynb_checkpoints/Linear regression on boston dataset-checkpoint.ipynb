# Linear Regression on Boston housing dataset
---

**1.1: Downloading the dataset**

```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_boston

%matplotlib inline

boston = load_boston()

boston_data = pd.DataFrame(boston.data, columns=boston.feature_names)
boston_data['MEDV'] = boston.target
boston_data.head()
```

---
**1.2 Information of the boston dataset**

```python
info = boston.DESCR
print(*info.split("**"))
```

* ***The dataset has 506 datapoint with 13 independent features and 1 dependent feature(medv)***
* ***We will use Linear regression algoritm to train a model to forecast house price given the features***

---
**1.3 Data preprocessing**

```python
# check if the data contians any null values

boston_data.isnull().sum() # no null values
```

---
**1.4 EDA**

```python
# lets do the corr relation calulation between variables to find the most useful variables
corr_matrix = boston_data.corr()

# heatmap of the corr matrix
plt.figure(figsize=(15,10))
sns.heatmap(data=corr_matrix, annot=True, cmap="YlGnBu")
```

The pearson correlation values range from -1 to 1, where -1 mean negative effect, 1 mean positive effect, while 0 means no relation

from the heatmap, The following features has a very good correlation with target variable

* LSTAT has -0.74 negative effect on target value, i.e prices
* B has 0.33 postive effect
* PTRATIO has -0.51 negative
* RM has 0.7 positive effect
* INDUS has 0.48 negative effect

***Note: The RAD and TAX have positive effect on each other, hence should not be choosen together in a model, same goes of DIS and AGE*** 

```python
plt.figure(figsize=(15,15))
sns.pairplot(data=boston_data[['LSTAT', 'B', 'PTRATIO', 'RM', 'INDUS', 'MEDV']])
```

* From the plots, the LSTAT and RM has better linearity with target variables
* we will train two models,
    * with all 5 feautres
    * with only LSTAT and RM

---
**1.5: Modelling**

```python
# model 1

X =  boston_data[['LSTAT', 'B', 'PTRATIO', 'RM', 'INDUS']]
Y = boston_data['MEDV']

# split the data into 80:20
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
print("X-train shape: ", x_train.shape)
print("Y-train shape: ", y_train.shape)
print("X-test shape: ", x_test.shape)
print("y-test shape: ", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

lin_model1 = LinearRegression()
lin_model1.fit(x_train, y_train)
y_train_predict = lin_model1.predict(x_train)

rmse = np.sqrt(mean_squared_error(y_train, y_train_predict))
r2 = r2_score(y_train, y_train_predict)

print()
print("The model performance for training set")
print(f'RMSE is {rmse}')
print(f'R2 score is {r2}')
print()

# test data
y_test_predict = lin_model1.predict(x_test)

rmse_t = np.sqrt(mean_squared_error(y_test, y_test_predict))
r2_t = r2_score(y_test, y_test_predict)

print()
print("The model performance for test set")
print(f'RMSE is {rmse_t}')
print(f'R2 score is {r2_t}')
print()
```

```python
# model 2

X =  boston_data[['LSTAT','RM', 'PTRATIO']]
Y = boston_data['MEDV']

# split the data into 80:20
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
print("X-train shape: ", x_train.shape)
print("Y-train shape: ", y_train.shape)
print("X-test shape: ", x_test.shape)
print("y-test shape: ", y_test.shape)
```

```python

lin_model2 = LinearRegression()
lin_model2.fit(x_train, y_train)
y_train_predict_2 = lin_model2.predict(x_train)

rmse2 = np.sqrt(mean_squared_error(y_train, y_train_predict))
r2_2 = r2_score(y_train, y_train_predict_2)

print()
print("The model performance for training set")
print(f'RMSE is {rmse2}')
print(f'R2 score is {r2_2}')
print()

# test data
y_test_predict_2 = lin_model2.predict(x_test)

rmse_t_2 = np.sqrt(mean_squared_error(y_test, y_test_predict_2))
r2_t_2 = r2_score(y_test, y_test_predict_2)

print()
print("The model performance for test set")
print(f'RMSE is {rmse_t_2}')
print(f'R2 score is {r2_t_2}')
print()
```
